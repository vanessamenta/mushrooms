```python
data.replace('?', np.nan , inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop("stalk-root", axis=1, inplace=True)
```

```python
s = (data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
label_data=data.copy()
label_encoder=LabelEncoder()
for col in object_cols:
    label_data[col]=label_encoder.fit_transform(label_data[col])
```

```python
label_data
```

```python
#separo target e predizioni e divido tra training e validation
from sklearn.model_selection import train_test_split


y=label_data.class_edible
X=label_data.drop("class_edible", axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=0)
print(X_train.shape, y_train.shape, X_valid.shape, y_valid.shape)
```

```python
y
```

```python
#Definisco il modello 
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

mushroom_model=DecisionTreeRegressor(random_state=1)
#Alleno il modello
mushroom_model.fit(X_train,y_train)

#validazione e calcolo del Mean absosulte error
val_pred=mushroom_model.predict(X_valid)

print(mean_absolute_error(y_valid,val_pred))
```

```python
#accuratezza
from sklearn.metrics import accuracy_score

acc_score = accuracy_score(val_pred.round(), y_valid) * 100
print("Accuracy Score:", acc_score)
```
